```python
# Program 16: Hierarchical Clustering

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Load dataset
iris = pd.read_csv("C:\\Users\\Uday Giri\\OneDrive\\Desktop\\IRIS.csv")

# Features and labels
x = iris.iloc[:, :-1].values
y = iris.iloc[:, -1].values

# Hierarchical clustering
hc = AgglomerativeClustering(n_clusters=3)
y_pred = hc.fit_predict(x)

# Plot the clusters (using first 2 features for visualization)
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s=50, c='blue', label='Cluster 3')

plt.title("Hierarchical Clustering (Agglomerative)")
plt.legend()
plt.show()

# Dendrogram
plt.figure(figsize=(8,5))
dendrogram(linkage(x, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Euclidean distance")
plt.show()
```
